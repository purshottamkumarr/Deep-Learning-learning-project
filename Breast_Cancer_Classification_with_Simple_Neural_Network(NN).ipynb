```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
```

```python
Breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
Breast_cancer_dataset
```

```python
data_frame = pd.DataFrame(Breast_cancer_dataset.data,columns = Breast_cancer_dataset.feature_names)
data_frame.head(5)
```

Adding the target Columns In the Dataframe

```python
data_frame["label"] =Breast_cancer_dataset.target
```

Separting the features and target

```python
x = data_frame.drop(columns ="label",axis =1)
y = data_frame["label"]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```


Standarize The data

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
x_train_std = Scaler.fit_transform(x_train)
x_test_std = Scaler.transform(x_test)
```

Building The Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

Setting up the layers of Neural Network

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(30,activation="relu"),
    keras.layers.Dense(2,activation = "sigmoid")
])
```

Compling the Neural Network

```python
model.compile(optimizer="adam",loss = "sparse_categorical_crossentropy",metrics =["accuracy"])
```

Training the Neural Network

```python
history = model.fit(x_train,y_train,validation_split=0.1,epochs=5)
```

Visualizing Accuracy and Loss

```python
plt.figure(figsize=(5,4))
plt.xlabel("Epochs",color="blue")
plt.ylabel("Accuracy",color="blue")
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Visulazing accuracy and Loss ")
plt.legend(["Train validation","Validation data"],loc = "lower right")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title("Visulazing Loss and val Loss")
plt.xlabel("Epochs",color = "red")
plt.ylabel("loss")
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.legend(["loss","val_loss"],loc = "upper right")
plt.show()
```

Accuracy Of the model On test Data

```python
loss,accuracy = model.evaluate(x_test_std,y_test)
print(accuracy)
```
